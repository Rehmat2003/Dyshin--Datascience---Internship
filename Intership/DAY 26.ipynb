```python
#load dataset

import pandas as pd
df=pd.read_csv("clean_movie_metadata.csv")
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

find the movies with the highest profit

```python
df[df["gross"] == df["gross"].max()][["movie_title","gross"]]
```

find IMDB rank with top 100

```python
df.sort_values(by="imdb_score", ascending=False).head(100)[["movie_title","imdb_score"]]
```

best director

```python
df.groupby("director_name")["imdb_score"].mean().sort_values(ascending=False).head(1)
```

popular genres

```python
df["genres"].value_counts().head()
```

critic favorite

```python
df.groupby("actor_1_name")["imdb_score"].mean().sort_values(ascending=False).head(1)
```

audience favorite

```python
df.groupby("actor_1_name")["num_voted_users"].sum().sort_values(ascending=False).head(1)
```
