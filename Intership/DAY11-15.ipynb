```python
#TO CHANGE THE DATATYPE

def update_commute_time(val):
    strips=['min','mins','minutes','hrs','hour']
    for substr in strips:
        if substr in val:
            split_index=val.find(substr)
            new_val=float(val[:split_index].strip())

            return new_val if substr =='min' else 60.0*new_val
        
    substr='apx.'
    if substr in val:
        split_index=val.find(substr)
        return float(val[split_index + len(substr):].strip())
        
    if "campus" in val:
        return 0
        
    return float(val)
```

```python
df["commute_mins"] = df["commute_time"].apply(update_commute_time)
```

```python
print(pd.unique(df.commute_mins))
#print(df.commute_mins)
```

```python
def update_height(val):
    substr = "cm"
    if substr in val:
        split_idx = val.find(substr)
        return float(val[:split_idx].strip())

    substr = "."
    if substr in val:
        if val.find(substr) == 1:
            return float(val) * 100.0
        return float(val)

    sunstr=","
    if substr in val:
        val = val.replace(substr,".")
        return float(val) * 100.0

    substr = "\'"
    if substr in val:
        split_strs = val.replace("\'"," ").replace("\""," ").split(" ")
        feet = float(split_strs[0])
        inches = float(split_strs[1])

        return 2.54 * inches + 30.48 * feet

        return float(val)
```

```python
df.height = df.height.apply(update_height)
print(pd.unique(df.height))
```
