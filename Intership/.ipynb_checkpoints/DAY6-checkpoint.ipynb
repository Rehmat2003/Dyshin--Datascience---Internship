```python
#check whether the number is even or odd using function
def even_odd(n):
    if n%2==0:
        print('even')
    else:
        print('odd')
n=int(input()) 
even_odd(n)
```

```python
#check whether number is positive or negative

def pos_neg(n):
    if n > 0:
        print("Positive Number")
    elif n < 0:
        print("Negative Number")
    else:
        print("Zero")

pos_neg(-1)
```

```python
#check which number is bigger

def bigger(a, b):
    if a > b:
        print(a, "is bigger")
    else:
        print(b, "is bigger")

a = int(input("Enter first number: "))

b = int(input("Enter second number: "))

bigger(a,b)
```

```python
#check for factorial

def fact(n):
    f=1
    for i in range(1, n + 1):
        f=f*i
    print("Factorial of", n, "is", f)

fact (4)
```

```python
#check for fibonacci

def fibonacci(n):
    a, b = 0, 1
    for i in range(n):
        print(a, end=" ")
        a, b=b, a + b

n=int(input("Enter number of terms: "))

fibonacci(n)
```

```python
#factorial using recursion

n=int(input("Enter the number"))
def fact(n):
    if n==1 or n==0:
        return 1
    else:
        return n*fact(n-1)
print("Factorial of", n, "is", fact(n))
```
