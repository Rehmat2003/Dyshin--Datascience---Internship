```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns 
```

# 2. Data Analysis

## 2a) Read the Dataset

**Explanation**

The first step in any data analysis project is to load the dataset into the analysis environment and inspect its structure. This allows us to verify column names, data types, and ensure the dataset has been correctly imported before performing any transformation or analysis.

```python
df = pd.read_csv(r"C:\Users\adity\Downloads\Dyashin Technosoft DS Internship\Datasets\US_births_2000-2014_SSA.csv")
df.head()
```

# 2b) Remove All Rows Corresponding to Holidays

**Explanation**

Certain days such as public holidays show unusually low birth counts due to hospital scheduling practices. These irregular observations can distort statistical analysis and visualizations. Removing these rows helps ensure the dataset reflects normal birth patterns.

```python
df_clean = df[df["births"] >= 10000]
```

# 2c) Remove All Weekday Rows Where Births Are Less Than 10,000

**Explanation**

Birth counts below 10,000 on weekdays usually indicate irregularities or holidays. Removing these ensures that weekday statistics represent typical hospital operations rather than exceptional cases.

```python
df_clean = df[(df["births"] >= 10000) & (df["day_of_week"] <= 5)]
```

# 2d) Average Number of Births Per Day for Each Day of the Week

**Explanation**

The average (mean) helps identify the general birth pattern for each weekday. This metric provides insight into whether births are evenly distributed throughout the week or concentrated on specific days.

**Insight**

The results typically show higher birth averages on **Tuesday, Wednesday, and Thursday**, suggesting hospitals schedule many births during mid-week.

```python
avg_births = df_clean.groupby("day_of_week")["births"].mean()
print(avg_births)
```

# 2e) Median Number of Births Per Day for Each Day of the Week

**Explanation**

The median represents the middle value of birth counts and is less affected by extreme values or outliers. Comparing the median with the mean helps understand whether the distribution is symmetric or skewed.


**Insight**

If the median is close to the mean, it indicates that the birth distribution across days is relatively stable.

```python
median_births = df_clean.groupby("day_of_week")["births"].median()
print(median_births)
```

# 2f) Dispersion of Births Per Day for Each Day of the Week

**Explanation**

Dispersion measures how much birth counts vary around the average. Standard deviation is commonly used to measure this variability. A higher value indicates more fluctuation in daily births.

**Insight**

Moderate dispersion indicates that daily birth counts fluctuate but generally stay within a predictable range.

```python
dispersion = df_clean.groupby("day_of_week")["births"].std()
print(dispersion)
```

# 2g) Interquartile Range (IQR) of Births Per Day

**Explanation**

The Interquartile Range (IQR) measures the spread of the middle 50% of the data by calculating the difference between the third quartile (Q3) and the first quartile (Q1). It is useful for understanding the variability in birth counts without being influenced by extreme values.

**Insight**

The IQR shows the normal spread of birth counts for each weekday and helps in identifying potential outliers.

```python
def calculate_iqr(x):
    q1 = x.quantile(0.25)
    q3 = x.quantile(0.75)
    return q3 - q1

iqr_values = df_clean.groupby("day_of_week")["births"].apply(calculate_iqr)
print(iqr_values)
```

# 2h) Identify Outliers Using the IQR Method

**Explanation**

Outliers are values that fall far outside the typical range of the dataset. Using the IQR method helps detect abnormal birth counts that may represent unusual events or reporting anomalies.

**The outlier detection rule is:**

Lower Bound = Q1 − 1.5 × IQR
Upper Bound = Q3 + 1.5 × IQR

**Insight**

Outliers often correspond to unusual events such as holidays, data recording anomalies, or special healthcare scheduling patterns.

```python
outliers = {}

for day, group in df_clean.groupby("day_of_week"):
    
    Q1 = group["births"].quantile(0.25)
    Q3 = group["births"].quantile(0.75)
    IQR = Q3 - Q1
    
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    
    outliers[day] = group[(group["births"] < lower) | (group["births"] > upper)]

for day in outliers:
    print(f"Day {day} outliers count:", len(outliers[day]))
```

# 2i) Box and Whisker Plot for Birth Distribution

**Explanation**

Box plots visually summarize the distribution of data by showing the median, quartiles, and potential outliers. They are useful for comparing birth distributions across different days of the week.

**Insight**

The plot typically shows higher births during weekdays and lower births during weekends.

```python
sns.boxplot(x="day_of_week", y="births", data=df_clean)
plt.title("Birth Distribution by Day of Week")
plt.show()
```

# 2j) Histogram and KDE Plot

**Explanation**

Histograms help visualize the frequency distribution of birth counts, while KDE (Kernel Density Estimation) provides a smooth curve representing the data distribution.

**Insight**

Birth counts generally follow a near-normal distribution with some variation due to healthcare scheduling patterns.

```python
sns.histplot(df_clean["births"], kde=True)
plt.title("Distribution of Daily Births")
plt.show()
```

# 2k) Day of the Year When 50% of Births Have Occurred

**Explanation**

This analysis identifies the point in the calendar year when 50% of the total annual births have occurred. By calculating the cumulative sum of daily births for each year and comparing it with half of the total yearly births, we determine the exact day of the year when the midpoint of births is reached. This helps assess whether births are evenly distributed throughout the year or concentrated during specific periods.

**Insight**

The results show that, in most years, the 50% birth threshold is reached around **mid-year (late June to early July)**. This suggests that births are generally distributed fairly evenly across the year, with no extreme seasonal concentration.

```python
df_clean = df_clean.rename(columns={"date_of_month":"day"})
df_clean["date"] = pd.to_datetime(df_clean[["year","month","day"]])
df_clean["day_of_year"] = df_clean["date"].dt.dayofyear
df_clean = df_clean.sort_values(["year","day_of_year"])
df_clean["cum_births"] = df_clean.groupby("year")["births"].cumsum()
total_births = df_clean.groupby("year")["births"].transform("sum")
result = df_clean[df_clean["cum_births"] >= total_births/2].groupby("year").first()
print(result["day_of_year"])
```

# 2l) In 2000, Between What Days Do 50% of Births Occur?

**Explanation**

This analysis focuses specifically on the year 2000 and identifies the range of days during which half of the year's births occurred. This helps understand the yearly birth distribution.

**Insight**

For the year 2000, approximately **half of the births occurred by early July**, indicating that births are relatively evenly distributed across the year.

```python
df_2000 = df_clean[df_clean["year"] == 2000]
df_2000["cumulative_births"] = df_2000["births"].cumsum()
half_births = df_2000["births"].sum() / 2
day_50_percent = df_2000[df_2000["cumulative_births"] >= half_births].iloc[0]
print(day_50_percent)
```
