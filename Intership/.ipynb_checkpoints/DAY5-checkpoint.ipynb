```python
#if stmt

n=30
if n>0:
    print ("My num is positive")
```

```python
#Else

age = 18

if age>=18:
    print('eligible to vote')
else:
    print('not eligible|')
```

```python
#elif

score = 85
if score>=90:
    print("A")
elif score>=70:
    print("B")
elif score>=40:
    print("C")
else:
    print("Fail")
```

```python
#match
day=3
match day:
    case 1:
        print('Monday')
    case 2:
        print('tuesday')
    case 3:
        print('wednesday')
    case 4:
        print('thursday')
    case 5:
        print('friday')
    case 6:
        print('saturday')
    case 7:
        print('Sunday')
    
```

```python
#nested if

marks=70
if marks>=35:
    print('pass')
    if marks>=90:
        print('Grade A')
    else:
        print('Grade B or C')
```

```python
#loops
#for i in range (1,11): this will also work
for i in range (10):
    print(i)
```

```python
for i in range (10,21,1):
    print(i)
```

```python
for i in range (20,10,-2):
    print(i)
```

```python
for i in 'apple':
    print(i)
```

```python
#while

i=0
while i<=5:
    print(i)
    i+=1
```

```python
#break

for i in range(5):
    if i==5:
        break
    print(i)
```

```python
#continue

for i in range(10):
    if i==5:
        continue
    print(i)
```
