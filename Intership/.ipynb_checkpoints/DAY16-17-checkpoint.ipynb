```python
#LOAD DATASET
import pandas as pd
df=pd.read_csv('US_births_2000-2014_SSA.csv')
df
```

```python
df.head()
```

```python
df.describe().sum()
```

```python
#null

df.isnull().sum()
```

```python
#manipulate

df[df["day_of_week"].isnull()]
```

```python
import numpy as np
```

```python
num_col = df.select_dtypes(include=np.number).columns
```

```python
cat_col = df.select_dtypes(exclude=np.number).columns
```

```python
for col in num_col:
    df[col].fillna(df[col].median(),inplace=True)

for col in cat_col:
    df[col].fillna(df[col].mode(),inplace=True)
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
#outliers - data points wch live extremely 

outlier_summary={}

for col in num_col:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3-Q1

    #limits

    lower = Q1-1.5*IQR
    upper = Q3+1.5*IQR

    #Outlier
    outlier = df[(df[col]<lower)|(df[col]>upper)]

    outlier_summary[col]=len(outlier)
```

```python
outlier_summary
```
