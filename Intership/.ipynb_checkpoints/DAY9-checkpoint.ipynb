```python
#basic function
def greet():
    print("Hello")

greet()
```

```python
#recursive function ex:1
def greet(n):
    #base function
    if n<=0:
        return
    #Recursive function    
    print("Hello")
    greet(n-1)
greet(4)
```

```python
#recursive function ex:2
def greet(n):
    #base function
    if n>=7:
        return
    #recursive function    
    print("Hello")
    greet(n+1)
greet(4)
```
