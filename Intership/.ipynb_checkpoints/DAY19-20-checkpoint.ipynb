### *Histogram plotting*

```python
import seaborn as sns
```

```python
sns.histplot(df['age'])
```

### *Bar graph*

```python
sns.barplot(df,x='age',y='city_name')
```

### *Pie chart*

```python
import matplotlib.pyplot as plt
x = df['STATUS_YEAR'].value_counts()
plt.pie(x.values, 
        labels = x.index,
        autopct = '%1.1f%%')

plt.show()
```

### *DAY 20*

```python
plt.figure(figsize=(15,8))
sns.barplot(x=df["department_name"],y=df["length_of_service"])
plt.xticks(rotation=45)
plt.show()
```

```python
sns.scatterplot(data=df,x="length_of_service",y="age")
```

```python
#categorical vs categorical
sns.countplot(data=df,x='department_name',hue='gender_full')
plt.xticks(rotation=90)
plt.show()
```

```python
sns.heatmap(df.select_dtypes(include=['number']).corr(),annot=True)
```
